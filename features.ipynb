## Finding Outliers
Using the Interquartile Range (IQR) method to identify outliers in numerical columns. The IQR is a measure of statistical dispersion and is used to detect outliers by calculating the range between the first quartile (Q1) and the third quartile (Q3).


```python
#identifying outliers using IQR method
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Exclude ID and boolean columns
bool_cols = [col for col in df.columns if np.all(np.isin(df[col].unique(), [0, 1]))]
numerical_cols = [col for col in numerical_cols if col not in bool_cols and col != 'ID']
# list of Outlier
def outlier_summary(dataframe):

    outlier_info = []
    total_outliers = 0

    for col in numerical_cols:
        # Calculate IQR
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1

        # Find outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
        outlier_count = len(outliers)

        if outlier_count > 0:
            outlier_percentage = (outlier_count / len(dataframe)) * 100
            outlier_info.append((col, outlier_count, outlier_percentage))
            total_outliers += outlier_count

    print(f"Columns with outliers: {len(outlier_info)}")
    print(f"Total outliers: {total_outliers}")

    if outlier_info:
        print("Columns with outliers:")
        print()
        # Sort by outlier count (descending)
        #outlier_info.sort(key=lambda x: x[1], reverse=True)

        for col, count, percentage in outlier_info:
            print(f"{col}: {count} outliers ({percentage:.1f}%)")
    else:
        print("No outliers detected in any column!")

# Run the outlier summary
outlier_summary(df)
```

```python
#visualize the outliers using boxplots
def plot_outliers(dataframe):
    n_cols = 3
    n_plots = len(numerical_cols)
    n_rows = (n_plots + n_cols - 1) // n_cols  # Ceiling division

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i, col in enumerate(numerical_cols):
        plt.subplot(n_rows, n_cols, i + 1)
        sns.boxplot(x=dataframe[col])
        plt.title(f'Boxplot of {col}')
        plt.xlabel(col)

    plt.tight_layout()
    plt.show()


# Run the outlier visualization
plot_outliers(df)
```

## handling the outliers
# We will use the capping method to handle outliers. This involves replacing outliers with the nearest non-outlier value, which is determined by the IQR method.


```python
# Capping method to handle outliers
def cap_outliers(dataframe):

    df_capped = dataframe.copy()

    for col in numerical_cols:
        # Calculate IQR
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Count outliers before capping
        outliers_before = len(df_capped[(df_capped[col] < lower_bound) | (df_capped[col] > upper_bound)])

        # Apply capping
        df_capped[col] = df_capped[col].clip(lower=lower_bound, upper=upper_bound)

        # Count outliers after capping
        outliers_after = len(df_capped[(df_capped[col] < lower_bound) | (df_capped[col] > upper_bound)])

        if outliers_before > 0:
            print(f"{col}: {outliers_before} outliers capped")

    print(f"\nOutlier capping completed!")
    return df_capped #return the capped DataFrame while preserving the original DataFrame

# Apply capping to the dataset
df_capped = cap_outliers(df)



```

```python
# Silhouette Score Comparison: Original vs Engineered Features (Fixed)

from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load both datasets
df_original = pd.read_csv('cleaned_customer_segmentation.csv')
df_featured = pd.read_csv('featured_customer_segmentation.csv')

print("=== SILHOUETTE SCORE COMPARISON ===\n")

# Define feature sets (ONLY NUMERIC FEATURES)
original_features = [
    'Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
    'Education', 'Recency', 'Kidhome', 'Teenhome'
]

engineered_features = [
 'Income', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'Age', 'Total_Dependents', 'Is_Parent'
]

# Function to calculate silhouette scores
def evaluate_clustering(data, features, name, k_range=range(2, 8)):
    print(f"\n{name} Features:")
    print(f"Features used: {features}")
    
    # Prepare data - only select numeric columns that exist
    available_features = [f for f in features if f in data.columns]
    print(f"Available features: {available_features}")
    
    X = data[available_features].copy()
    
    # Handle missing values for numeric columns only
    for col in X.columns:
        if X[col].dtype in ['int64', 'float64']:
            X[col] = X[col].fillna(X[col].mean())
        else:
            print(f"Warning: Skipping non-numeric column {col}")
            X = X.drop(columns=[col])
    
    if X.empty or len(X.columns) == 0:
        print(f"Error: No valid numeric features for {name}")
        return [], [], 0, 0
    
    print(f"Final features for clustering: {list(X.columns)}")
    
    # Scale features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    silhouette_scores = []
    inertias = []
    
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(X_scaled)
        
        # Calculate silhouette score
        sil_score = silhouette_score(X_scaled, cluster_labels)
        silhouette_scores.append(sil_score)
        inertias.append(kmeans.inertia_)
        
        print(f"K={k}: Silhouette Score = {sil_score:.4f}")
    
    # Find optimal k
    best_k = k_range[np.argmax(silhouette_scores)]
    best_score = max(silhouette_scores)
    
    print(f"Best K: {best_k} with Silhouette Score: {best_score:.4f}")
    
    return silhouette_scores, inertias, best_k, best_score

# Evaluate original features
original_sil, original_inertia, orig_best_k, orig_best_score = evaluate_clustering(
    df_original, original_features, "ORIGINAL"
)

# Evaluate engineered features  
engineered_sil, engineered_inertia, eng_best_k, eng_best_score = evaluate_clustering(
    df_featured, engineered_features, "ENGINEERED"
)

# Calculate improvement (only if both evaluations succeeded)
if orig_best_score > 0 and eng_best_score > 0:
    improvement = ((eng_best_score - orig_best_score) / orig_best_score) * 100

    print(f"\n{'='*60}")
    print("FEATURE ENGINEERING IMPACT SUMMARY")
    print(f"{'='*60}")
    print(f"Original Best Score:    {orig_best_score:.4f} (K={orig_best_k})")
    print(f"Engineered Best Score:  {eng_best_score:.4f} (K={eng_best_k})")
    print(f"Improvement:            {improvement:+.2f}%")
    print(f"{'='*60}")

    # Visualization
    if len(original_sil) > 0 and len(engineered_sil) > 0:
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
        k_range = range(2, 8)
        
        # Silhouette scores comparison
        axes[0].plot(k_range, original_sil, 'o-', label='Original Features', linewidth=2, markersize=8)
        axes[0].plot(k_range, engineered_sil, 's-', label='Engineered Features', linewidth=2, markersize=8)
        axes[0].set_xlabel('Number of Clusters (K)')
        axes[0].set_ylabel('Silhouette Score')
        axes[0].set_title('Silhouette Score Comparison')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        # Highlight best scores
        axes[0].scatter([orig_best_k], [orig_best_score], color='blue', s=100, marker='*', 
                       label=f'Best Original (K={orig_best_k})')
        axes[0].scatter([eng_best_k], [eng_best_score], color='red', s=100, marker='*',
                       label=f'Best Engineered (K={eng_best_k})')
        
        # Inertia comparison  
        axes[1].plot(k_range, original_inertia, 'o-', label='Original Features', linewidth=2, markersize=8)
        axes[1].plot(k_range, engineered_inertia, 's-', label='Engineered Features', linewidth=2, markersize=8)
        axes[1].set_xlabel('Number of Clusters (K)')
        axes[1].set_ylabel('Inertia')
        axes[1].set_title('Inertia Comparison (Lower is Better)')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

    print(f"\nKEY IMPROVEMENTS FROM FEATURE ENGINEERING:")
    print(f"{'='*60}")
    print(f"âœ“ Added family-related features (Total_Dependents, Is_Parent)")
    print(f"âœ“ Created spending efficiency metrics (Average_Spent)")
    print(f"âœ“ Included customer lifecycle info (Age, Tenure_Days)")
    print(f"âœ“ Added channel preference ratios (Web/Store Purchase Ratios)")
    print(f"âœ“ Campaign response indicators (Ever_Accepted_Campaign)")
    print(f"âœ“ Consolidated spending metrics (Total_Spending, Total_Purchases)")

else:
    print("Error: Could not complete comparison due to feature issues")
```

```python


# Visualization of Feature Engineering Results

import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Load the featured dataset
df = pd.read_csv('featured_customer_segmentation.csv')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("Dataset shape:", df.shape)
print("\nCreated features:")
new_features = ['Age', 'Total_Dependents', 'Is_Parent', 'Total_Spending', 
               'Total_Purchases', 'Average_Spent', 'Tenure_Days', 
               'WebPurchaseRatio', 'StorePurchaseRatio', 'Ever_Accepted_Campaign']
print(new_features)

## 1. HISTOGRAMS FOR KEY DISTRIBUTIONS
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
fig.suptitle('Distribution of Engineered Features', fontsize=16, y=0.98)

# Key features to plot
features_to_plot = [
    'Age', 'Total_Dependents', 'Total_Spending', 'Total_Purchases',
    'Average_Spent', 'Tenure_Days', 'WebPurchaseRatio', 'StorePurchaseRatio',
    'Income', 'Education', 'Is_Parent', 'Ever_Accepted_Campaign'
]

for idx, feature in enumerate(features_to_plot):
    row = idx // 4
    col = idx % 4
    ax = axes[row, col]
    
    # Plot histogram with KDE
    sns.histplot(data=df, x=feature, kde=True, ax=ax, alpha=0.7)
    ax.set_title(f'Distribution of {feature}')
    ax.grid(True, alpha=0.3)
    
    # Add statistics
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    ax.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='--', label=f'Median: {median_val:.1f}')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()

## 2. 3D SCATTER PLOTS
# 3D Plot 1: Income vs Total_Spending vs Age (colored by Total_Dependents)
fig = plt.figure(figsize=(15, 5))

# First 3D plot
ax1 = fig.add_subplot(131, projection='3d')
scatter1 = ax1.scatter(df['Income'], df['Total_Spending'], df['Age'], 
                      c=df['Total_Dependents'], cmap='viridis', alpha=0.6, s=20)
ax1.set_xlabel('Income')
ax1.set_ylabel('Total Spending')
ax1.set_zlabel('Age')
ax1.set_title('Income vs Spending vs Age\n(Color: Dependents)')
plt.colorbar(scatter1, ax=ax1, shrink=0.5)

# Second 3D plot: Age vs Tenure vs Total_Purchases (colored by Education)
ax2 = fig.add_subplot(132, projection='3d')
scatter2 = ax2.scatter(df['Age'], df['Tenure_Days'], df['Total_Purchases'], 
                      c=df['Education'], cmap='plasma', alpha=0.6, s=20)
ax2.set_xlabel('Age')
ax2.set_ylabel('Tenure (Days)')
ax2.set_zlabel('Total Purchases')
ax2.set_title('Age vs Tenure vs Purchases\n(Color: Education)')
plt.colorbar(scatter2, ax=ax2, shrink=0.5)

# Third 3D plot: Income vs Average_Spent vs WebPurchaseRatio (colored by Is_Parent)
ax3 = fig.add_subplot(133, projection='3d')
scatter3 = ax3.scatter(df['Income'], df['Average_Spent'], df['WebPurchaseRatio'], 
                      c=df['Is_Parent'], cmap='coolwarm', alpha=0.6, s=20)
ax3.set_xlabel('Income')
ax3.set_ylabel('Average Spent')
ax3.set_zlabel('Web Purchase Ratio')
ax3.set_title('Income vs Avg Spent vs Web Ratio\n(Color: Is Parent)')
plt.colorbar(scatter3, ax=ax3, shrink=0.5)

plt.tight_layout()
plt.show()

## 3. SCATTER PLOT MATRIX FOR KEY RELATIONSHIPS
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Key Feature Relationships', fontsize=16)

# Plot 1: Income vs Total_Spending (colored by Age)
scatter1 = axes[0,0].scatter(df['Income'], df['Total_Spending'], 
                            c=df['Age'], cmap='viridis', alpha=0.6, s=30)
axes[0,0].set_xlabel('Income')
axes[0,0].set_ylabel('Total Spending')
axes[0,0].set_title('Income vs Total Spending\n(Color: Age)')
plt.colorbar(scatter1, ax=axes[0,0])

# Plot 2: Age vs Total_Dependents (colored by Total_Spending)
scatter2 = axes[0,1].scatter(df['Age'], df['Total_Dependents'], 
                            c=df['Total_Spending'], cmap='plasma', alpha=0.6, s=30)
axes[0,1].set_xlabel('Age')
axes[0,1].set_ylabel('Total Dependents')
axes[0,1].set_title('Age vs Dependents\n(Color: Spending)')
plt.colorbar(scatter2, ax=axes[0,1])

# Plot 3: Tenure vs Total_Purchases (colored by Income)
scatter3 = axes[0,2].scatter(df['Tenure_Days'], df['Total_Purchases'], 
                            c=df['Income'], cmap='coolwarm', alpha=0.6, s=30)
axes[0,2].set_xlabel('Tenure (Days)')
axes[0,2].set_ylabel('Total Purchases')
axes[0,2].set_title('Tenure vs Purchases\n(Color: Income)')
plt.colorbar(scatter3, ax=axes[0,2])

# Plot 4: WebPurchaseRatio vs StorePurchaseRatio (colored by Age)
scatter4 = axes[1,0].scatter(df['WebPurchaseRatio'], df['StorePurchaseRatio'], 
                            c=df['Age'], cmap='viridis', alpha=0.6, s=30)
axes[1,0].set_xlabel('Web Purchase Ratio')
axes[1,0].set_ylabel('Store Purchase Ratio')
axes[1,0].set_title('Web vs Store Purchase Ratio\n(Color: Age)')
plt.colorbar(scatter4, ax=axes[1,0])

# Plot 5: Education vs Average_Spent (colored by Total_Dependents)
scatter5 = axes[1,1].scatter(df['Education'], df['Average_Spent'], 
                            c=df['Total_Dependents'], cmap='plasma', alpha=0.6, s=30)
axes[1,1].set_xlabel('Education Level')
axes[1,1].set_ylabel('Average Spent per Purchase')
axes[1,1].set_title('Education vs Average Spent\n(Color: Dependents)')
plt.colorbar(scatter5, ax=axes[1,1])

# Plot 6: Income vs Age (colored by Ever_Accepted_Campaign)
scatter6 = axes[1,2].scatter(df['Income'], df['Age'], 
                            c=df['Ever_Accepted_Campaign'], cmap='coolwarm', alpha=0.6, s=30)
axes[1,2].set_xlabel('Income')
axes[1,2].set_ylabel('Age')
axes[1,2].set_title('Income vs Age\n(Color: Campaign Response)')
plt.colorbar(scatter6, ax=axes[1,2])

plt.tight_layout()
plt.show()

## 4. BOX PLOTS FOR CATEGORICAL COMPARISONS
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Feature Distributions by Categories', fontsize=16)

# Box plot 1: Total_Spending by Is_Parent
sns.boxplot(data=df, x='Is_Parent', y='Total_Spending', ax=axes[0,0])
axes[0,0].set_title('Spending by Parental Status')
axes[0,0].set_xticklabels(['No Children', 'Has Children'])

# Box plot 2: Income by Education Level
sns.boxplot(data=df, x='Education', y='Income', ax=axes[0,1])
axes[0,1].set_title('Income by Education Level')

# Box plot 3: Average_Spent by Total_Dependents
sns.boxplot(data=df, x='Total_Dependents', y='Average_Spent', ax=axes[0,2])
axes[0,2].set_title('Average Spent by Number of Dependents')

# Box plot 4: WebPurchaseRatio by Ever_Accepted_Campaign
sns.boxplot(data=df, x='Ever_Accepted_Campaign', y='WebPurchaseRatio', ax=axes[1,0])
axes[1,0].set_title('Web Purchase Ratio by Campaign Response')
axes[1,0].set_xticklabels(['No Response', 'Responded'])

# Box plot 5: Total_Purchases by Age Groups
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 100], labels=['Young', 'Middle', 'Mature', 'Senior'])
sns.boxplot(data=df, x='Age_Group', y='Total_Purchases', ax=axes[1,1])
axes[1,1].set_title('Purchases by Age Group')

# Box plot 6: Tenure_Days by Income Quartiles
df['Income_Quartile'] = pd.qcut(df['Income'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
sns.boxplot(data=df, x='Income_Quartile', y='Tenure_Days', ax=axes[1,2])
axes[1,2].set_title('Tenure by Income Quartile')

plt.tight_layout()
plt.show()

## 5. CORRELATION HEATMAP OF NEW FEATURES
plt.figure(figsize=(12, 10))
correlation_features = ['Age', 'Total_Dependents', 'Total_Spending', 'Total_Purchases', 
                       'Average_Spent', 'Tenure_Days', 'WebPurchaseRatio', 
                       'StorePurchaseRatio', 'Income', 'Education']

corr_matrix = df[correlation_features].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix of Key Features', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("FEATURE ENGINEERING SUMMARY")
print("="*50)
print(f"Total customers: {len(df):,}")
print(f"Age range: {df['Age'].min():.0f} - {df['Age'].max():.0f} years")
print(f"Average family size: {df['Total_Dependents'].mean():.1f} dependents")
print(f"Parents: {df['Is_Parent'].sum():,} ({df['Is_Parent'].mean()*100:.1f}%)")
print(f"Average spending: ${df['Total_Spending'].mean():.0f}")
print(f"Average tenure: {df['Tenure_Days'].mean():.0f} days")
print(f"Campaign responders: {df['Ever_Accepted_Campaign'].sum():,} ({df['Ever_Accepted_Campaign'].mean()*100:.1f}%)")
```
